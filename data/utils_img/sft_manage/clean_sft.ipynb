```python
import json

input_file = "/home/vcj9002/jianshu/chengxuan/ProgressLM/data/train/visual_demo/visual_h5_franka_3rgb_rl.jsonl"
output_file = "/home/vcj9002/jianshu/chengxuan/ProgressLM/data/train/visual_demo/visual_franka_3rgb_new_sft.jsonl"

with open(input_file, 'r') as f:
    data = [json.loads(line) for line in f]

print(f"筛选前总数: {len(data)}")

filtered_data = [item for item in data if 3 <= int(item['total_steps']) <= 7]

print(f"筛选后总数: {len(filtered_data)}")
print(f"筛选掉: {len(data) - len(filtered_data)}")

with open(output_file, 'w') as f:
    for item in filtered_data:
        f.write(json.dumps(item, ensure_ascii=False) + '\n')

print(f"已保存到: {output_file}")
```

```python
import json

input_file = "/home/vcj9002/jianshu/chengxuan/ProgressLM/data/train/visual_demo/visual_franka_3rgb_n.jsonl"
output_file = "/home/vcj9002/jianshu/chengxuan/ProgressLM/data/train/visual_demo/visual_franka_3rgb_new.jsonl"

with open(input_file, 'r') as f:
    data = [json.loads(line) for line in f]

for item in data:
    item['data_source'] = item['data_source'].replace('robomind_', '')

with open(output_file, 'w') as f:
    for item in data:
        f.write(json.dumps(item, ensure_ascii=False) + '\n')

print(f"已处理 {len(data)} 条记录")
print(f"输出文件: {output_file}")
```

```python
import json

input_file = "/home/vcj9002/jianshu/chengxuan/ProgressLM/data/train/rl/text_rl_all.jsonl"
output_file = "/home/vcj9002/jianshu/chengxuan/ProgressLM/data/train/rl/text_rl_all_3to7.jsonl"

with open(input_file, 'r') as f:
    data = [json.loads(line) for line in f]

for item in data:
    item['data_source'] = item['data_source'].replace('robomind_', '')
    item['total_steps'] = str(item['total_steps'])

with open(output_file, 'w') as f:
    for item in data:
        f.write(json.dumps(item, ensure_ascii=False) + '\n')

print(f"已处理 {len(data)} 条记录")
print(f"输出文件: {output_file}")
```
