```python
import json

input_file = "/projects/p32958/chengxuan/ProgressLM/data/train/rl/visual_rl_all.jsonl"
output_file = "/projects/p32958/chengxuan/ProgressLM/data/train/rl/visual_rl_3to7.jsonl"

with open(input_file, 'r') as f:
    data = [json.loads(line) for line in f]

print(f"筛选前总数: {len(data)}")

filtered_data = [item for item in data if 3 <= int(item['total_steps']) <= 7]

print(f"筛选后总数: {len(filtered_data)}")
print(f"筛选掉: {len(data) - len(filtered_data)}")

with open(output_file, 'w') as f:
    for item in filtered_data:
        f.write(json.dumps(item, ensure_ascii=False) + '\n')

print(f"已保存到: {output_file}")
```

```python
import json

input_file = "/projects/p32958/chengxuan/ProgressLM/data/train/text_demo/new/edited_nega_text_sftt.jsonl"
output_file = "/projects/p32958/chengxuan/ProgressLM/data/train/text_demo/new/edited_nega_text_sft.jsonl"


with open(input_file, "r", encoding="utf-8") as fin, open(output_file, "w", encoding="utf-8") as fout:
    for line in fin:
        if not line.strip():
            continue
        data = json.loads(line)
        # 尝试将字符串转为数字
        if "closest_idx" in data:
            try:
                data["closest_idx"] = int(data["closest_idx"])
            except ValueError:
                try:
                    data["closest_idx"] = float(data["closest_idx"])
                except ValueError:
                    pass  # 如果不是数字字符串则保持原样
        fout.write(json.dumps(data, ensure_ascii=False) + "\n")

```

```python
import json

# 输入输出路径
src_path = "/projects/p32958/chengxuan/ProgressLM/data/train/text_demo/new/edited_nega_text_sftt.jsonl"      # 要调整顺序的文件
ref_path = "/projects/p32958/chengxuan/ProgressLM/data/train/text_demo/new/new_text_negative_sft_raw.jsonl"   # 目标顺序的参考文件
out_path = "/projects/p32958/chengxuan/ProgressLM/data/train/text_demo/new/edited_nega_text_sft.jsonl"      # 输出文件

# 读取参考顺序
with open(ref_path, "r", encoding="utf-8") as ref_file:
    ref_order = [json.loads(line)["id"] for line in ref_file]

# 读取源文件并建成字典
with open(src_path, "r", encoding="utf-8") as src_file:
    src_data = {json.loads(line)["id"]: line.strip() for line in src_file}

# 按参考顺序输出
with open(out_path, "w", encoding="utf-8") as out_file:
    for _id in ref_order:
        if _id in src_data:
            out_file.write(src_data[_id] + "\n")

print("✅ 已按参考文件顺序重新排序并输出到：", out_path)
```
