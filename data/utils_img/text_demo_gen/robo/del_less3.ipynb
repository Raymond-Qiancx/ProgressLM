```python
import json

def filter_and_report(input_file, output_file):
    total_count = 0
    kept_count = 0
    removed_count = 0
    
    with open(input_file, 'r', encoding='utf-8') as infile, \
         open(output_file, 'w', encoding='utf-8') as outfile:
        
        for line in infile:
            total_count += 1
            data = json.loads(line)
            
            if data.get('total_steps', 0) >= 3:
                outfile.write(json.dumps(data, ensure_ascii=False) + '\n')
                kept_count += 1
            else:
                removed_count += 1
    
    print(f"总记录数: {total_count}")
    print(f"保留记录: {kept_count}")
    print(f"删除记录: {removed_count}")

filter_and_report('/home/vcj9002/jianshu/workspace/data/COCO/annotations/raw/text/h5_agilex_3rgb_text.jsonl', '/home/vcj9002/jianshu/workspace/data/COCO/annotations/raw/text/h5_agilex_3rgb_text_all.jsonl')
```
